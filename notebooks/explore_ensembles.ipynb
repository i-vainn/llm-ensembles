### Mode

```python
from itertools import combinations
import statistics as stat


def evaluate_ensembles_mode(res):
    acc = 0
    for pred, gt in zip(map(stat.mode, res), data):
        acc += (gt['target_index'] == pred)
    return(acc / len(data))

results = []
for n in [2, 3, 4]:
    for comb in combinations(paths[:4], n):
        predicts = []
        for path in comb:
            df = pd.read_json(path)
            predicts.append(df.acc_norm_ans_ctx.apply(np.array).values)
        res = None
        for predict in predicts:
            cur = np.array(list(map(np.argmax, predict))).reshape(-1, 1)
            if res is None:
                res = cur
            else:
                res = np.hstack([res, cur])
        print("Models: {}".format(list(map(lambda x: x[:-22], comb))))
        print("Score: {:.4f}\n".format(evaluate_ensembles_mode(res)))
```

### Simple logit averaging

```python
def evaluate_ensembles(res):
    acc = 0
    for pred, gt in zip(res, data):
        acc += (gt['target_index'] == pred)
    return(acc / len(data))
```

```python
results = []
for n in [2, 3, 4]:
    for comb in combinations(paths[:4], n):
        predicts = []
        for path in comb:
            df = pd.read_json(path)
            predicts.append(df.acc_norm_ans_ctx.apply(np.array).values)
        res = predicts[0]
        for predict in predicts[1:]:
            for i in range(len(predict)):
                res[i] += predict[i]
        answer = map(np.argmax, res)
        print("Models: {}".format(list(map(lambda x: x[:-22], comb))))
        print("Score: {:.4f}\n".format(evaluate_ensembles(answer)))
```
